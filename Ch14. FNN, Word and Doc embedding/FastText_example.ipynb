```python
with open('2016_filtered_review.txt', encoding='utf-8') as f:
    docs = [line.strip().split('\t\t') for line in f]

docs1 = [doc[1].strip().split() for doc in docs]
```

```python
print(docs1[:3])
```

```python
from gensim.models import FastText
```

```python
model_sg_n10 = FastText(docs1, window=3, min_count=3, vector_size=100, sg=1, negative=10, 
                        min_n=2, max_n=5)
```

```python
model_sg_n10.wv.most_similar("이정재", topn=50)
```
