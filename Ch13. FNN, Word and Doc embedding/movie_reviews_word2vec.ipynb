```python
from gensim.models import Word2Vec
```

```python
with open('2016_filtered_review.txt', encoding='utf-8') as f:
    docs = [line.strip().split('\t\t') for line in f]
docs1 = [doc[1].strip().split() for doc in docs]
```

```python
print(docs1[:3])
```

### sg=1, negative=10 인 경우

```python
model_sg_n10 = Word2Vec(docs1, window=3, min_count=3, vector_size=100, sg=1, negative=10)
# https://radimrehurek.com/gensim/models/word2vec.html
```

```python
model_sg_n10.wv['이정재']
```

```python
len(model_sg_n10.wv['이정재'])
```

```python
import numpy as np
```

```python
print(np.dot(model_sg_n10.wv['이정재'], model_sg_n10.wv['이범수'])/(np.linalg.norm(model_sg_n10.wv['이정재'])*np.linalg.norm(model_sg_n10.wv['이범수'])))
```

```python
model_sg_n10.wv.similarity('이정재', '이범수')
```

```python
model_sg_n10.wv.similar_by_word('이정재',topn=50)
```

```python
model_sg_n10.wv.similar_by_word('재밌',topn=50)
```
