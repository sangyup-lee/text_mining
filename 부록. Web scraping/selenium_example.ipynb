```python
import time

from selenium import webdriver # selenium의 webdriver 모듈을 임포트 합니다.
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

# chromedriver.exe 파일의 경로를 입력합니다.
s = Service(r'C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe')
driver = webdriver.Chrome(service=s)

url = 'http://www.yes24.com'
driver.get(url) 
time.sleep(2)
# '베스트' 링크 찾아서 클릭하기
element = driver.find_element(by=By.XPATH, value='//*[@id="yesFixCorner"]/dl/dd/ul[1]/li[1]/a')
element.click()

time.sleep(2)
# 베스트셀러 페이지의 첫 번째 책 링크 찾고 클릭하기
element = driver.find_element(by=By.XPATH, value='//*[@id="bestList"]/ol/li[1]/p[1]/a')
element.click()

time.sleep(2)
# 회원 리뷰 링크 찾고 클릭하기
element = driver.find_element(by=By.XPATH, value='//*[@id="yDetailTopWrap"]/div[2]/div[1]/span[3]/span[2]/a')
element.click()

html = driver.page_source # 해당 페이지의 source code 실시간으로 다운로드 받기
time.sleep(1)

driver.close()
```

```python
# 셀레니엄을 이용해 다운로드된 소스코드 확인하기
print(html)
```
