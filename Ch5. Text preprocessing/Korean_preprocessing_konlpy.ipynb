```python
with open('Korean_news.txt', 'r', encoding='utf8') as f:
    content = f.read()
```

```python
print(content)
```

```python
# 기호 제거하기
import re
filtered_content = re.sub(r'[^\d\s\w]',' ', content)
```

```python
print(filtered_content)
```

```python
import konlpy.tag
```

```python
komoran = konlpy.tag.Komoran()
komoran_morphs = komoran.pos(filtered_content)
print(komoran_morphs)
```

```python
# 품사 태그 확인
komoran.tagset
```

```python
print(komoran.nouns(filtered_content))
```

```python
# Twitter has changed to Okt since 0.4.5
okt= konlpy.tag.Okt()
okt_morphs = okt.pos(filtered_content)
print(okt_morphs)
```

```python
okt.tagset
```

```python
Noun_words = okt.nouns(filtered_content)
print(Noun_words)
```

```python
Featured_POS = ['Noun', 'Adjective', 'Verb']
selected_words = []
for word, pos in okt_morphs:
    if pos in Featured_POS:
        selected_words.append(word)
print(selected_words)
```

```python
# 불용어 제거를 위한 별도의 사전 (list 변수) 구축
stopwords = ['연합뉴스', '기자', '다음', '이것', '저것']
unique_Noun_words = set(Noun_words)
for word in unique_Noun_words:
    if word in stopwords or len(word)==1:
        while word in Noun_words: Noun_words.remove(word)
```
